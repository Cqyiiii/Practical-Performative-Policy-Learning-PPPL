```python
import torch 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns;sns.set()
import json
from matplotlib.ticker import FuncFormatter
```

```python
seq_seed_list = range(20, 30)
T = 300
epoch = list(range(T)) * len(seq_seed_list)

value_vanillaGD = []
value_strategicGD = []

for seed in seq_seed_list:
    value_vanillaGD.extend(torch.load("result/vani_seed{}.pkl".format(seed)))    
    value_strategicGD.extend(torch.load("result/stra_seed{}.pkl".format(seed)))
    
with open('result/cutoff.json', 'r') as file:
    value_cutoff = json.load(file)    
```

```python
df = pd.DataFrame({'epoch': epoch, 'value_strategicGD': value_strategicGD, 'value_vanillaGD': value_vanillaGD})
```

```python
def mul_100(x, pos):
    return f'{x*100}k'

plt.figure(figsize=(15, 10))

sns.lineplot(data = df, x = 'epoch', y = 'value_strategicGD', label='Strategic Policy Gradient',  errorbar=('ci', 95))
sns.lineplot(data = df, x = 'epoch', y = 'value_vanillaGD', label='Vanilla Policy Gradient', errorbar=('ci', 95))
# sns.lineplot(data = df, x = 'epoch', y = 'value_end2endGD', label='End2end Policy Gradient', color='green', errorbar=('ci', 95))
plt.axhline(y=sum(value_cutoff) / len(value_cutoff), color='red', linestyle='--', label='Policy: CATE>0')

plt.gca().yaxis.set_major_formatter(FuncFormatter(mul_100))
plt.xlabel("number of epoch", fontsize=20)
plt.ylabel("policy value", fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=30)
plt.tight_layout()
plt.savefig('semi-performance.pdf')
plt.show()
```

```python
T = 300
for col in df.columns[1:]:
    best_value = []
    for rep in range(10):
        best_value.append(df.loc[rep*T:(rep+1)*T, col].max())
    best_value = np.array(best_value)
    print(col, best_value.mean(), best_value.std())
```
